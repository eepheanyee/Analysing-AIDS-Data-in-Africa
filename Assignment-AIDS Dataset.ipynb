**Finding the country with highest number of HIV infected people**
For this, we have to find the maximum value in the `Data.People Living with HIV.Total` column

```python
df = pd.DataFrame(aids)
```

```python
df
```

I'll be using the `Data.People Living with HIV.Total` column  be cause from the dtaa description, that column gives us the total number of people suffering from AIDS in this year (The latest year in the data set)

```python

column = df["Data.People Living with HIV.Total"]
print(column)
max_value = column.max()
column_index = column.idxmax()
print(max_value)
print(column_index)
```

I want to check the row to confirm this

```python
print(df.iloc[column_index])
```

Now we have ascertained the highest number in the `Data.People Living with HIV.Total` column and it's index, we can use the index to get the country from the country column

```python
country = df["Country"]
```

```python
country
```

```python
highest_country = country[column_index]
```

```python
highest_country
```

From this, we can tell that South Africa has the highest number of people living with AIDS.
